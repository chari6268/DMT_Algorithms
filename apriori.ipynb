## Transaction Encoder

```python
from mlxtend.preprocessing import TransactionEncoder
```

## apriori and association_rules

```python
from mlxtend .frequent_patterns import apriori,association_rules
```

## Using and 'TransactionEncoder' object, we can transform this dataset into an array format suitable for typical machine learning APIs. Via the 'fit' method, the TransactionEncoder learns the unique labels in the dataset, and via the transform method, it transforms the input dataset (a Python list of lists) into a one-hot encoded NumPy boolean array

```python
transactions=[]
items=data['Item Names'].values
# print(items)
for i in range (0,len(items)):
    transactions.append(items[i].split(","))
print(transactions)
```

```python
TE=TransactionEncoder()
trans=TE.fit(transactions).transform(transactions)
print(trans)
```

## After fitting, the unique column names that correspond to the data array shown above can be accessed via the columns_ attribute:

```python
dataTE=pd.DataFrame(data = trans, columns = TE.columns_, dtype = int)
print(dataTE)
```

```python
dataTE.head()
```

## Support refers to the default popularity of an item and can be calculated by finding number of transactions containing a particular item divided by total number of transactions.

```python
support=apriori(dataTE,min_support=0.2,use_colnames=True)
support.sort_values(by = 'support', ascending = True)
```

Example:
    Support(BISCUIT) = (6)/(19) == 0.315789

confidence (x->y) = Support(x U y) / Support(x)

```python
confidence = association_rules(support, metric = 'confidence', min_threshold = 0.2)
confidence.sort_values(by = 'confidence', ascending = True)
```
